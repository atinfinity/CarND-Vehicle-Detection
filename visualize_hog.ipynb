# HOG Visualization

## import package

```python
import cv2
import matplotlib.pyplot as plt
```

```python
import vehicleDetectUtil as vehicleUtil
import vehicleDetect_svmVar as svmVar
```

## HOG Visualization(vehicle)

### read vehicle image

```python
image = cv2.imread('vehicles/vehicles/KITTI_extracted/1.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

### extract HOG feature

```python
# Call our function with vis=True to see an image output
features, hog_image = vehicleUtil.get_hog_features(gray, orient= 9, 
                        pix_per_cell= 8, cell_per_block= 2, 
                        vis=True, feature_vec=False)
```

### visualization

```python
# Plot the examples
fig = plt.figure()
plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Example Vehicle Image')
plt.subplot(122)
plt.imshow(hog_image, cmap='gray')
plt.title('HOG Visualization')
```
